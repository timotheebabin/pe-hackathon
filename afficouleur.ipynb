```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
dicouleur={1:[255,255,255],
           2:[255,0,0],
           3:[0,255,0],
           4:[0,0,255],
           5:[255,255,0],
           6:[255,0,255],
           7:[0,255,255],
           8:[0,0,0],
           9:[125,0,255],
           10:[255,125,0],
           11:[0,255,125],
           12:[125,255,0],
           13:[255,0,125],
           14:[0,125,255]}

def afficouleur(tab):
    t_n,t_m=np.shape(tab)
    tabl=np.zeros(shape=(t_n,t_m,3))
    for i in range(t_n):
        for j in range(t_m):
            tabl[i][j]=dicouleur[tab[i][j]]
    plt.imshow(tabl)

afficouleur(np.array([[1,3,6],[3,2,8]]))
```
